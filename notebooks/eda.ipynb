# Exploratory Data Analysis (EDA)

This notebook is for understanding the data only - no model training yet.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/features.csv")

print(df.shape)
df.head()
```

```python
missing = df.isnull().mean().sort_values(ascending=False)

print(missing)
```

# Column Categorization

```python
electron1_cols = [
    "E1", "px1", "py1", "pz1",
    "pt1", "eta1", "phi1", "Q1"
]

electron2_cols = [
    "E2", "px2", "py2", "pz2",
    "pt2", "eta2", "phi2", "Q2"
]
```

# Create Missing Flags

```python
for col in df.columns:
    if col != "id":
        df[f"{col}_missing"] = df[col].isnull().astype(int)

df.head()
```

# Physics Features

```python
# Reconstruct Transverse Momentum
df["pt1_reco"] = np.sqrt(df["px1"]**2 + df["py1"]**2)
df["pt2_reco"] = np.sqrt(df["px2"]**2 + df["py2"]**2)

# Compare Measured vs Reconstructed
df["pt1_diff"] = abs(df["pt1"] - df["pt1_reco"])
df["pt2_diff"] = abs(df["pt2"] - df["pt2_reco"])
```

```python
# Reconstruct Energy
df["E1_reco"] = np.sqrt(
    df["px1"]**2 +
    df["py1"]**2 +
    df["pz1"]**2
)

df["E2_reco"] = np.sqrt(
    df["px2"]**2 +
    df["py2"]**2 +
    df["pz2"]**2
)
```

```python
# Angular Features
df["delta_phi"] = abs(df["phi1"] - df["phi2"])
df["delta_eta"] = abs(df["eta1"] - df["eta2"])
df["delta_R"] = np.sqrt(
    df["delta_eta"]**2 +
    df["delta_phi"]**2
)
```

```python
# Event-Level Features
df["total_pt"] = df["pt1"] + df["pt2"]
df["total_E"] = df["E1"] + df["E2"]
df["charge_product"] = df["Q1"] * df["Q2"]
```

# Save Processed Features

```python
df.to_csv(
    "../data/processed/features_engineered.csv",
    index=False
)
```
